```python
#Problem 2
board1 = [
['X','X','O'],
['O','X','O'],
['O','','X'],
]

board2 = [
['X','X','O'],
['O','X','O'],
['','O','X'],
]

board3 = [
['O','X','O'],
['','O','X'],
['X','X','O'],
]

board4 = [
['X','X','X'],
['O','X','O'],
['O','','O'],
]

board5 = [
['X','X','O'],
['O','X','O'],
['X','','O'],
]

board6 = [
['X','X','O'],
['O','X','O'],
['X','',''],
]

def count(line):          #function to help count all x's and o's in the given list
    O_count = [element.count("O") for element in line]
    X_count = [element.count("X") for element in line]
    return O_count, X_count

def get_winner(board):
    #rows
    row = [row for row in board]
    O_row, X_row = count(row)
    
    #columns
    column = [a for a in zip(*board)]
    O_column, X_column = count(column)

    #diagonals (neg and pos are the "slopes" hahahaha)
    diagonal_neg = [board[i][i] for i in range(0,3)]
    diagonal_pos = [board[i][2-i] for i in range(0,3)]
    
    #added both diagonals to one list
    diagonals = []
    diagonals.append(diagonal_neg)
    diagonals.append(diagonal_pos)
    O_diagonals, X_diagonals = count(diagonals)
    
    
    #win conditions... this assumes there are no "defective boards"
    #defective (ie. there are 3 X's and 3 O's in a straight line... so "both should win"...)
    #but there should never be a case like that in the first place (hence, "defective")
    #this code would return "O" if there was a defective board since return "O" is above return "X"
    if 3 in O_row or 3 in O_column or 3 in O_diagonals:
        return "O"
    elif 3 in X_row or 3 in X_column or 3 in X_diagonals:
        return "X"
    else:
        return "No Winner"
    
try:
    print(get_winner(board5))
except:
    pass
```

```python
#Problem 3

locations = {
    "admu":{
        "name":"Ateneo de Manila University",
        "latitude":14.6394557,
        "longitude":121.0758975        
    },
    "upd":{
        "name":"University of the Philippines-Diliman Campus",
        "latitude":14.6538519,
        "longitude":121.0663348
    },
    "dlsu":{
        "name":"De La Salle University-Taft",
        "latitude":14.5647346,
        "longitude":120.9909876

    },
    
}

legs = {
     ("upd","admu"):{
         "travel_time_mins":10
     },
     ("admu","dlsu"):{
         "travel_time_mins":35
     },
     ("dlsu","upd"):{
         "travel_time_mins":55
     }
}

def eta(source,destination):
    current = source
    time = 0
    while current != destination:
        for element in legs.keys():
            if current == element[0]:
                time += legs[element]["travel_time_mins"]
                current = element[1]
                break
    return ('ETA: %d' % time)


source = input("Enter source: ")
while source not in locations:           #just to make sure source is within the given locations dict above
    print("Invalid input!")
    source = input("Enter source: ")
    

destination = input("Enter destination: ")    #just to make sure destination is within the given locations dict above
while destination not in locations:
    print("Invalid input!")
    destination = input("Enter destination: ")


print(eta(source,destination))
```
