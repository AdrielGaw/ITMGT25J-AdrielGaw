```python
# Write code here for People per Birthmonth
# Source: https://stackoverflow.com/questions/30009948/how-to-reorder-indexed-rows-based-on-a-list-in-pandas-data-frame
monthlist = ["January","February","March","April","May","June","July","August","September","October","November","December"]

df_birthmonth = recweek.copy().groupby(["birthmonth"]).count()
df_birthmonth = df_birthmonth.reindex(monthlist)
df_birthmonth.drop(columns = df_birthmonth.iloc[:,0:7],inplace = True)
df_birthmonth.reset_index(inplace=True)
df_birthmonth.columns = ["birthmonth","count"]
df_birthmonth
```

```python
# Sample output
# same day
# sort by calendar day
```

```python
# write code here for People per BirthDay

df_birthday = recweek.copy().groupby(["birthday"]).count()
df_birthday.drop(columns = df_birthday.iloc[:,0:7],inplace = True)
df_birthday.reset_index(inplace = True)
df_birthday.columns = ["birthday","count"]
df_birthday.sort_values(by="birthday",inplace=True)
df_birthday
```

```python
# Sample output
# same year
# sort accordingly
```

```python
# write code here for People per BirthYear

df_birthyear = recweek.copy().groupby(["birthyear"]).count()
df_birthyear.drop(columns = df_birthyear.iloc[:,0:7],inplace = True)
df_birthyear.reset_index(inplace = True)
df_birthyear.columns = ["birthyear","count"]
df_birthyear.sort_values(by="birthyear",inplace=True)
df_birthyear
```

```python
# write code here for People per Yearlevel

df_yearlevel = recweek.copy().groupby(["yearlevel"]).count()
df_yearlevel.drop(columns = df_yearlevel.iloc[:,0:7],inplace = True)
df_yearlevel.reset_index(inplace = True)
df_yearlevel.columns = ["yearlevel","count"]
df_yearlevel.sort_values(by="yearlevel",inplace=True)
df_yearlevel
```

```python
# write code here for People per Course

df_course = recweek.copy().groupby(["course"]).count()
df_course.drop(columns = df_course.iloc[:,0:7],inplace = True)
df_course.reset_index(inplace = True)
df_course.columns = ["course","count"]
df_course.sort_values(by="course",inplace=True)
df_course
```

```python
# write code here for People per Carrier+Gender
# Source: https://www.geeksforgeeks.org/python-pandas-series-to_frame/
# Source: https://stackoverflow.com/questions/33346591/what-is-the-difference-between-size-and-count-in-pandas
prefixes_dict = {"0917":"Globe","0927":"Globe","0918":"Smart"}
remove_columns = ["firstname","lastname","birthmonth","birthday","birthyear","course","yearlevel"]

carrier = recweek.copy()
carrier.drop(columns = remove_columns,inplace=True)
carrier.columns = ["gender",'carrier']
carrier["carrier"].replace(r" [0-9]{3} [0-9]{4}","",regex=True,inplace=True)
carrier["carrier"].replace(prefixes_dict,inplace=True)
```

```python
# Series output
series = carrier.groupby(["carrier","gender"]).size()
series
```

```python
# Dataframe output
df_carrier_gender = series.to_frame()
df_carrier_gender.reset_index(inplace=True)
df_carrier_gender.columns = ["carrier","gender","count"]
df_carrier_gender
```

```python
# Sample output
# carrier count
```

```python
# write code here for People per Carrier

df_carrier = carrier.copy()
df_carrier.drop(columns = "gender",inplace=True)

df_carriercount = df_carrier.groupby(["carrier"]).size().to_frame()
df_carriercount.reset_index(inplace=True)
df_carriercount.columns = ["carrier","count"]
df_carriercount
```
